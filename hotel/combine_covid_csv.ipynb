```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
tf.keras.utils.set_random_seed(42)
```

```python
stem = "/Volumes/Seagate Portable Drive/Unpaired_COVID/"

filelist = os.listdir(stem)
print(len(filelist))
filelist.remove("bulk_download2.sh")
filelist.remove("Light_Chains")
filelist.remove(".DS_Store")
# filelist.remove("total.csv")
print(len(filelist))
```

```python
# import shutil
# for i in filelist:
#     if i.find("Light") != -1:
#         shutil.move(stem + i, stem + "Light_Chains/" + i)
```

```python
for i in filelist:
    df = pd.read_csv(stem + i, skiprows=1)
    df_new = df[["sequence", "sequence_alignment", "sequence_alignment_aa", "Redundancy"]]
    df_new.to_csv(stem + i, index=False)
    print("done")
```

```python
df_append = pd.DataFrame()
#append all files together
for file in filelist:
    df_temp = pd.read_csv(stem + file)
    df_temp = df_temp[df_temp["Redundancy"].apply(lambda x: x >= 2)]
    df_append = df_append.append(df_temp, ignore_index=True)
df_append
```

```python
df_append.to_csv(stem + "total.csv", index=False)
```
