```python
def binomial(p, n, successes):
    cumulative_prob = 0
    for j in range(successes, n+1):
        cumulative_prob += (comb(n, j) * p**j * (1-p)**(n - j))
    return cumulative_prob
```

```python
healthy_array = []
for i in range(1, 40):
    prob = 0.8493
    healthy_array.append(binomial(prob, i, math.floor(i * 0.5)))
```
